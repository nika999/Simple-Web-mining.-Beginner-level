```python
import requests
from requests import get
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from time import sleep
from random import randint
```

```python
titles =[]
companies =[]
pages = np.arange(1, 12,1)
```

```python
for page in pages:
    page = requests.get("https://djinni.co/jobs/keyword-data-analyst/?page=" + str(page))
    soup = BeautifulSoup(page.text, 'html.parser')
    job_h2 = soup.find_all('a', class_="profile")
    company = soup.find_all('span', class_="public-salary-item")
    
    sleep(randint(2,10))
    
    for link2 in company:
        name2 = link2.get_text()
        companies.append(name2)
    
```

```python
salary = list(map(str.strip,companies))
print(salary)
```

```python
df = pd.DataFrame({'Salaries':salary})

df = df.replace('\$', '', regex=True)
df
```

```python
df[['From','To']] = df.Salaries.str.split("-",expand=True,)
df
```

```python
print(df[['From']].mean(axis=0))
```
