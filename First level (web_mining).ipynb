Here we will unload data from the https://pudding.cool/ which will contain title, description and author of articles.

We will use Pandas for data manipulating, Urllib3 is used to open URLs and the Beautiful Soup package is used to extract data from html files.

```python
import pandas as pd
import urllib3
import bs4
from bs4 import BeautifulSoup
http = urllib3.PoolManager()
r = http.request('GET', 'https://pudding.cool/')
r.status
# 200 HTTP Status Codes means OK
```

Here we collect all the data that is on this page:

```python
d = r.data
d
```

And create nested data structure document:

```python
soup = BeautifulSoup(d, 'html.parser')
sp =soup.prettify()
print(soup.prettify())
```

After studying the structure we will find titles of articles in this way:

```python
soup.find_all('h3')
```

Clean data and create list of titles:

```python
title =[]
for link in soup.find_all('h3'):
    name = link.get_text()
    print(link.get_text())
    title.append(name)
```

We can study structure better by pulling out pieces of structure:

```python
soup.main.li.div
```

```python
authors =[]
for link in soup.find_all('cite'):
    name2 = link.get_text()
    print(link.get_text())
    authors.append(name2)
```

```python
subtitle =[]
for link in soup.find_all('p', class_='item__dek'):
    name3 = link.get_text()
    print(link.get_text())
    subtitle.append(name3)
```

Here we are:

```python
df = pd.DataFrame({'Title name':title,'Author':authors, 'Subtitle':subtitle})
df
```
